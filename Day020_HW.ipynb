```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
Tr = table.find_all('tr')
air_con = ({})
count = 0
for tr in Tr:
    for td in tr.find_all('td'):        
        try:
            if(td['style'] == 'white-space:nowrap;'):
                #print(td.text)
                count = 0
                row = dict({})
                air_con.update({td.text: row})
        except:
            count += 1
            if (count%3 == 2):
                key = td.text
            if (count%3 == 0):
                value = td.text
                #print(key, ":", value)
                row.update({key: value})
            continue
print(air_con)
```
