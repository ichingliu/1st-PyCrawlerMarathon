### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
res = requests.get(url)

soup = BeautifulSoup(res.text, "html5lib")
post = []
#<h3><span class="date">2019/12/15 23:29</span><em class="tag c_travel">旅遊</em><a href="/news/20191215/1505758.htm" target="_blank">機上觸犯法律會如何？空中警察可執行任何「解除危機」方法</a></h3>

for s in soup.find(class_="part_list_2").find_all('h3'):
    print(s.find(class_="date").text, s.find("em").text, s.find("a").text)
    post.append([s.find(class_="date").text, s.find("em").text, s.find_all("a")[-1].text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
from collections import Counter
cate = []
for p in post:
    cate.append(p[1])

Counter(cate)
```
